```python
import pandas as pd
from pyproj import Proj, transform
```

```python
df = pd.read_csv('/Users/juhong/Library/CloudStorage/OneDrive-아주대학교/대학교/3학년 2학기/캡스톤/직접 연계 통계 모음/모음/신호등/A057_L_부착대/A057_L.csv',encoding = 'utf-8')
```

```python
df.head()
```

```python
#위도,경도로 변환 EPSG5186 -> EPSG4326
proj_1 = Proj(init = 'epsg:5186')
proj_2 = Proj(init = 'epsg:4326')
```

```python
x_list = df['XCE']
y_list = df['YCE']
```

```python
converted = transform(proj_1,proj_2,x_list,y_list)
converted
```

```python
df['XCE'] = converted[0]
df['YCE'] = converted[1]
df.head()
```

```python
df.rename(columns = {"XCE":"위도","YCE":"경도"},inplace = True)
df.head()
```
